```python
# IMPORTANT, READ BELOW 



# it kinda gets weird here, there is no dataset available for "county-wise yearly dwelling construction on CSO's website" so we use the NDA05 dwelling constrctions by LEA. 
# most names in the list slready have the name of their county at the end, but some places like "Fingal" do not, hence a list of such keywords is defined and later merged into the dublin count
# Not being well versed with Ireland, the margin for error is definitely present. Though it has been thoroughly checked witht he help of agents like claude and gemini, errors still might exist.
   # also claude really helped with the coding for this bit ;)
#________________________________________________________________________________________________________________________________________________________________________________________________#


# Define subregions that belong to Dublin (Fingal, South Dublin, Dún Laoghaire-Rathdown, Dublin City)
dublin_keywords = ['Dublin City', 'Fingal', 'South Dublin', 'Dún Laoghaire', 'Dun Laoghaire']

# Extract counties from LEA name
def extract_counties(electoral_area):
    # Handle Dublin sub-counties
    if any(keyword in electoral_area for keyword in dublin_keywords):
        return ['Dublin']
    # Handle the rest
    matches = []
    for county in counties:
        if county in electoral_area:
            matches.append(county)
    return matches if matches else None

# Add 'Counties' column
NDA05['Counties'] = NDA05['Local Electoral Area'].apply(extract_counties)

# Expand rows for multiple matches
expanded_rows = []
for _, row in NDA05.iterrows():
    if row['Counties'] is not None:
        for county in row['Counties']:
            new_row = row.copy()
            new_row['County'] = county
            expanded_rows.append(new_row)


expanded_df = pd.DataFrame(expanded_rows)

# filtering by counties
CNDA05 = expanded_df[expanded_df['County'].isin(counties)]

# group and summaiizng 
total_data = CNDA05.groupby(['Year', 'County'])['VALUE'].sum().reset_index()
total_data = total_data.rename(columns={'VALUE': 'Total Dwellings Built'})
results_CNDA05 = total_data.sort_values(['Year', 'County'])



display(results_CNDA05)
```

```python
house_size = NDA08[NDA08['Statistic Label'] == 'Average New Dwelling Size']
weight_mix = NDA08[NDA08['Statistic Label'] == 'Weight in Mix']


#  merging into single df on year on type of dwelling
NDA08_merged = pd.merge(
    house_size,
    weight_mix,
    on=['Year', 'Dwelling Type'],
    suffixes=('_size', '_mix')
)



# converting the % into numbers
NDA08_merged['Mix Fraction'] = NDA08_merged['VALUE_mix'] / 100

# contribution of each home type into the mix
NDA08_merged['Weighted Size'] = NDA08_merged['VALUE_size'] * NDA08_merged['Mix Fraction']


# final average sizes by year.
weighted_avg_per_year = NDA08_merged.groupby('Year')['Weighted Size'].sum().reset_index()
weighted_avg_per_year.rename(columns={'Weighted Size': 'Weighted Avg Dwelling Size (sqm)'}, inplace=True)

print(weighted_avg_per_year)

# we can see a downtrend post 2012
```

```python

#merging on year & county
County_DwellingInfo = pd.merge(results_CNDA05, weighted_avg_per_year, on='Year', how='left')

# estimated total floor area, (avg. dwelling size is just there.)
County_DwellingInfo["Estd. Total Floor Area (sqm)"] = (
    County_DwellingInfo['Total Dwellings Built'] * County_DwellingInfo['Weighted Avg Dwelling Size (sqm)']
)

display(County_DwellingInfo)
```

```python

# filtering chpa 02 further to only keep desireed fields
CHPA02_filtered = (
    CHPA02[
        (CHPA02['Type of Buyer'] == 'Household Buyer - All') &
        (CHPA02['Type of Sale'] == 'Market')
    ]
    .groupby(['Year', 'County'], as_index=False)['Mean Sale Price']
    .mean()
)

#filtering nea 08 further
CNEA08_filtered = CNEA08[CNEA08['Sex'] == 'Both sexes'][['Year', 'County', 'Median Annual Earnings']]


# merging to get our final dataset
master_df = CHPA02_filtered.copy()

master_df = master_df.merge(CNEA08_filtered, on=['Year', 'County'], how='left')
master_df = master_df.merge(results_CNDA05, on=['Year', 'County'], how='left')
master_df = master_df.merge(
    County_DwellingInfo[['Year', 'County', 'Estd. Total Floor Area (sqm)', 'Weighted Avg Dwelling Size (sqm)']],
    on=['Year', 'County'],
    how='left'
)

# Price/m2 calculation (just division)
master_df['Price per sqm'] = (
    master_df['Mean Sale Price'] /
    master_df['Weighted Avg Dwelling Size (sqm)']
).round(2)

# sorting by county and year
final_master_df = master_df.sort_values(['County', 'Year']).reset_index(drop=True)

# preview
final_master_df

#Printing final dataset to desired location, (remember to un-comment)
#final_master_df.to_csv("Expanded DF.csv", index=False)
```
